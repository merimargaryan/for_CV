```python
from sklearn.feature_extraction.text import CountVectorizer
import pymorphy2
import random
import Levenshtein
import pandas as pd
import nltk
from nltk import word_tokenize
nltk.download()
```

```python
import csv
import nltk

# загрузка данных из файла
with open('preprocessed_descriptions.csv', 'r', newline='', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    # собираем все описания рецептов
    descriptions = [row['descriptions'] for row in reader]

# токенизация текста
tokens = []
for description in descriptions:
    tokens.extend(nltk.word_tokenize(description))

# получение набора уникальных слов
words = set(tokens)
```

1.3 Напишите функцию, которая для заданного слова `word` возвращает `k` ближайших к нему слов из списка `words` (близость слов измеряется с помощью расстояния Левенштейна)

```python
def find_closest_words(word, words, k):
    # Получаем расстояния Левенштейна между словами и word
    distances = [Levenshtein.distance(word, w) for w in words]
    # Создаем список кортежей (word, distance)
    word_distances = list(zip(words, distances))
    # Сортируем список по возрастанию расстояний Левенштейна
    sorted_distances = sorted(word_distances, key=lambda x: x[1])
    # Получаем k ближайших слов
    result_words = [w[0] for w in sorted_distances[:k]]
    return result_words

word = 'butter'
k = 3
result = find_closest_words(word, words, k)
print(f'Ближайшие {k} слова к слову "{word}": {result}')
```
